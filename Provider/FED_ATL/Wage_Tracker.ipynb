```python
import pandas as pd
```

```python
url = 'https://www.atlantafed.org/-/media/documents/datafiles/chcs/wage-growth-tracker/wage-growth-data.xlsx'
df = pd.read_excel(url, 
                     sheet_name='Average Wage Quartile', 
                     header=2, 
                     index_col=0,
                     usecols=[0,1,2,3,4]
                    )

df = df.replace('.','NaN')
df = df.astype(float)
df = df.dropna()
df.head(2)
```

```python
df.plot()
```
