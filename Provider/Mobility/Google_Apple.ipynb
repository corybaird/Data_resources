## 1.1 Apple
- https://covid19.apple.com/mobility

```python
os.listdir('Input/')
df_mob_app = pd.read_csv('Input/applemobilitytrends-2022-01-30.csv')
df_mob_app.head(2)
```

### 1.1.1 California

```python
mask = (df_mob_app['sub-region'] == 'California') & (df_mob_app.region == 'Los Angeles')
df_mob_ca = df_mob_app.loc[mask, :].drop(['geo_type', 'region', 'alternative_name', 'sub-region','country'], axis=1)
df_mob_ca = df_mob_ca.pivot_table(columns ='transportation_type')
df_mob_ca.head(3)
```

### 1.1.2 US

```python
df_mob_app = df_mob_app.loc[df_mob_app.region == 'United States', :]
df_mob_app = df_mob_app.drop(['geo_type', 'region', 'alternative_name', 'sub-region','country'], axis=1)
df_mob_app.head(2)
```

```python
df_mob_app.melt(id_vars='transportation_type')
df_mob_app = df_mob_app.pivot_table(columns ='transportation_type')
df_mob_app.tail(3)
```
