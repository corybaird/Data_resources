```python
import pandas as pd
```

```python
url = 'https://www.mlit.go.jp/common/001461206.xlsx'
df_house = pd.read_excel(
    url,
    header = 5,
    index_col = 0).iloc[3: ,:]
df_house.head(2)
```

```python
df_house.dtypes
```

```python
df_house.index = pd.to_datetime(df_house.index)
df_house = df_house.astype(float)
df_house.head(3)
```

```python
df_house.dtypes
```

```python
df_house.pct_change(12).multiply(100).rolling(12).mean().dropna().loc[
    '2014':, ['Detached House', 'Condominiums']]
```
